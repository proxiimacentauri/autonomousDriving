```python
import pickle
import numpy as np
from numpy.linalg import inv
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from rotations import angle_normalize, rpy_jacobian_axis_angle, skew_symmetric, Quaternion
```

```python
#### 1. Data ###################################################################################

with open('data/pt1_data.pkl', 'rb') as file:
    data = pickle.load(file)

gt = data['gt']  # This is the Ground Truth
imu_f = data['imu_f'] # This is the Specific Forces from the IMU
imu_w = data['imu_w'] # This is the Rotational Rates from the IMU
gnss = data['gnss'] # This is the GNSS data
lidar = data['lidar'] # This is the LIDAR data
```

```python
                        """Data Dictionary"""
    
################################################################################################
# Each element is stored as an item from the data dictionary, which we
# will store in local variables, described by the following:
#
                        """ Ground Truth - (Intertial Frame) """
#     gt: Data object containing ground truth.
#     
#      With the following fields:
        #     a: Acceleration of the vehicle
        #     v: Velocity of the vehicle
        #     p: Position of the vehicle
        #     alpha: Rotational acceleration of the vehicle
        #     w: Rotational velocity of the vehicle
        #     r: Rotational position of the vehicle, in Euler (XYZ) angles
        #     _t: Timestamp in ms
#        
                        """ IMU - (Vehicle Frame)"""
#   imu_f: StampedData object with the imu specific force data
        #     data: The actual data
        #     t: Timestamps in ms
        
#   imu_w: StampedData object with the imu rotational velocity
        #     data: The actual data
        #     t: Timestamps in ms
        
                        """ GNSS """
#   gnss: StampedData object with the GNSS data.
        #     data: The actual data
        #     t: Timestamps in ms
#
                        """ LIDAR """
#   lidar: StampedData object with the LIDAR data (positions only).
        #     data: The actual data
        #     t: Timestamps in ms
################################################################################################
```

```python
                        """Modify this for Part - 2 (Extrinsic Calibration)"""
################################################################################################
# 1. LIDAR data is just a set of positions estimated from a separate scan-matching system.
#    So we can insert it into our solver as another position measurement similar to GNSS.

# 2. LIDAR frame is not the same as the frame as IMU and the GNSS. 

# 3. We transform the LIDAR data to the IMU frame using extrinsic calibration:

#     a.) Rotation Matrix C_li 
#     b.) Translation Vector t_i_li
################################################################################################

"""  Calibration of Rotation Matrix  """

# Correct calibration rotation matrix, corresponding to Euler RPY angles (0.05, 0.05, 0.1).
C_li = np.array([
   [ 0.99376, -0.09722,  0.05466],
   [ 0.09971,  0.99401, -0.04475],
   [-0.04998,  0.04992,  0.9975 ]
])

# Incorrect calibration rotation matrix, corresponding to Euler RPY angles (0.05, 0.05, 0.05).
# C_li = np.array([
#      [ 0.9975 , -0.04742,  0.05235],
#      [ 0.04992,  0.99763, -0.04742],
#      [-0.04998,  0.04992,  0.9975 ]
# ])

"""  Translation Vector  """

t_i_li = np.array([0.5, 0.1, 0.5])

""" Transform from the LIDAR frame to the vehicle (IMU) frame """

lidar.data = (C_li @ lidar.data.T).T + t_i_li
```

```python
#### 2. Constants ##############################################################################

"""  Setting Estimated Sensor Variances  """

var_imu_f = 0.10
var_imu_w = 0.25
var_gnss  = 0.01  # 0.06 for Part2
var_lidar = 1.00  # 88 for Part2

"""  Gravity  """
g = np.array([0, 0, -9.81]) 

"""  Motion Model Noise Jacobian  """
L_k = np.zeros([9, 6])
L_k[3:, :] = np.eye(6)

"""  Measurement Model Jacobian  """
H_k = np.zeros([3, 9])
H_k[:, :3] = np.eye(3)
```

```python
#### 3. Initial Values #########################################################################

Total_Time_Steps = imu_f.data.shape[0]

"""  Initial Position  """
p_est = np.zeros([Total_Time_Steps, 3]) 
p_est[0] = gt.p[0]

"""  Initial Velocity  """
v_est = np.zeros([Total_Time_Steps, 3])
v_est[0] = gt.v[0]

"""  Initial Orientation  """
# Stored as  Quaternion
q_est = np.zeros([Total_Time_Steps, 4])  
q_est[0] = Quaternion(euler=gt.r[0]).to_numpy()

""" Initial State Vector Co-Variance for every time-step """
P_k = np.zeros([Total_Time_Steps, 9, 9])
P_k[0] = np.zeros(9)  


""" Other Constants """
GNSS_Counter  = 0
LIDAR_Counter = 0
```

```python
"""    Measurement Update for GNSS and the LIDAR Data    """

def measurement_update(sensor_var, p_cov_check, y_k, p_check, v_check, q_check):
    
    """" 1. Compute Kalman Gain """
           
    # Sensor Noise Co-Variance (R_k) (GNSS or LIDAR)
    R_k = np.identity(3) * sensor_var
    
    K_k = p_cov_check.dot(H_k.T).dot(inv(H_k.dot(p_cov_check).dot(H_k.T) + R_k))

    """" 2. Compute Error State (delta_x) """
        
    delta_x = K_k.dot(y_k - p_check)
    
    """" 3. Correct the Predicted Error State (delta_x) """
    delta_p = delta_x[:3]  # Position is in Row 1 to 3
    delta_v = delta_x[3:6]  # Velocity is in Row 3 to 6
    delta_phi = delta_x[6:]  # Orientation is in Row 6 to 9
    
    Corrected_p = p_check + delta_p
    Corrected_v = v_check + delta_v
    Corrected_q = Quaternion(euler=delta_phi).quat_mult_right(q_check)
    

    """" 4. Compute the Corrected Co-Variance """
    
    p_cov_hat = (np.identity(9) - K_k.dot(H_k)).dot(p_cov_check)

    return Corrected_p, Corrected_v, Corrected_q, p_cov_hat
```

```python
                """############### Main Filter Loop ###############"""

################################################################################################
# Taking in the sensor data and creating estimates
################################################################################################

Total_LIDAR_Measurments_Avail = lidar.t.shape[0]
Total_GNSS_Measurments_Avail = gnss.t.shape[0]
Total_Time_Steps = imu_f.data.shape[0]

for k in range(1, Total_Time_Steps):  # We start at t = 1 because initial prediction from ground truth is taken
    T = imu_f.t[k] - imu_f.t[k - 1]
    
    Force_IMU = imu_f.data
    Omega_IMU = imu_w.data
    
    
    """" 1. Update the State with the IMU Measurement """
    
    Rotation_Mat = Quaternion(*q_est[k-1]).to_mat()
    
    """ 1.1 Update States (p, v, q) with IMU Readings (f, omega) from pickle file (Using Full Motion Model) """
    
    p_est[k] = p_est[k-1] + T * v_est[k-1] + ((T**2)/2) * (Rotation_Mat.dot(Force_IMU[k-1]) + g)
    
    v_est[k] =  v_est[k-1] + T * (Rotation_Mat.dot(Force_IMU[k-1]) + g)
    
    q_est[k] = Quaternion(axis_angle = (Omega_IMU[k-1]*T)).quat_mult_right(q_est[k-1])
    
    
    """ 1.2 Linearised Motion Model (F_k) """
    
    F_k = np.identity(9)
    
    # Row (1 to 3) Column (3 to 6)
    F_k[:3, 3:6] = T * np.identity(3)
    # Row (3 to 6) Column (6 to 9) .reshape((3,1))
    F_k[3:6, 6:] = -T * (Rotation_Mat.dot(skew_symmetric(Force_IMU[k-1].reshape((3,1)))))
       
    """ L_k is defined above """
    
    """ 2. Updating the State Co-Variance/Uncertianty (P_k) """
    
    # Process Noise Co-Variance (Q_k)
    Q_k = np.identity(6)
    # Row (1 to 3) Column (3 to 6)
    Q_k[:, :3] *= (T**2) * var_imu_f
    Q_k[:,-3:] *= (T**2) * var_imu_w
    
    P_k[k] = F_k.dot(P_k[k-1]).dot(F_k.T) + L_k.dot(Q_k).dot(L_k.T)

    """ Based on availability of GNSS and LIDAR measurements Calculate K_k, State Vector and Co-Variance"""
    
    
    if (LIDAR_Counter < Total_LIDAR_Measurments_Avail) and lidar.t[LIDAR_Counter] == imu_f.t[k-1]:
        
        p_est[k], v_est[k], q_est[k], P_k[k] = measurement_update(var_lidar, P_k[k], lidar.data[LIDAR_Counter].T, p_est[k], v_est[k], q_est[k])
        LIDAR_Counter += 1
        
    if (GNSS_Counter < Total_GNSS_Measurments_Avail) and gnss.t[GNSS_Counter] == imu_f.t[k-1]:
        p_est[k], v_est[k], q_est[k], P_k[k] = measurement_update(var_gnss, P_k[k], gnss.data[GNSS_Counter].T, p_est[k], v_est[k], q_est[k])
        GNSS_Counter += 1

    

    # Update states (save)
```

```python
#### 6. Results and Analysis ###################################################################

################################################################################################
# Now that we have state estimates for all of our sensor data, let's plot the results. This plot
# will show the ground truth and the estimated trajectories on the same plot. Notice that the
# estimated trajectory continues past the ground truth. This is because we will be evaluating
# your estimated poses from the part of the trajectory where you don't have ground truth!
################################################################################################
est_traj_fig = plt.figure(figsize=(12,10))
ax = est_traj_fig.add_subplot(111, projection='3d')
ax.plot(p_est[:,0], p_est[:,1], p_est[:,2], label='Estimated')
ax.plot(gt.p[:,0], gt.p[:,1], gt.p[:,2], label='Ground Truth')
ax.set_xlabel('Easting [m]', fontsize=16,labelpad=10)
ax.set_ylabel('Northing [m]', fontsize=16,labelpad=10)
ax.set_zlabel('Up [m]', fontsize=16,labelpad=10)
ax.set_title('Ground Truth and Estimated Trajectory', fontsize=16)
ax.set_xlim(0, 200)
ax.set_ylim(0, 200)
ax.set_zlim(-2, 2)
ax.set_xticks([0, 50, 100, 150, 200])
ax.set_xticklabels([0, 50, 100, 150, 200], fontsize=14)
ax.set_yticks([0, 50, 100, 150, 200])
ax.set_yticklabels([0, 50, 100, 150, 200], fontsize=14)
ax.set_zticks([-2, -1, 0, 1, 2])
ax.set_zticklabels([-2, -1, 0, 1, 2], fontsize=14)
ax.legend(loc=(0.62,0.77))
ax.view_init(elev=45, azim=-50)
plt.tight_layout(pad=5)
plt.show()

################################################################################################
# We can also plot the error for each of the 6 DOF, with estimates for our uncertainty
# included. The error estimates are in blue, and the uncertainty bounds are red and dashed.
# The uncertainty bounds are +/- 3 standard deviations based on our uncertainty (covariance).
################################################################################################
error_fig, ax = plt.subplots(2, 3,figsize=(16,8))
error_fig.suptitle('Error Plots', fontsize=20)
num_gt = gt.p.shape[0]
p_est_euler = []
p_cov_euler_std = []

# Convert estimated quaternions to euler angles
for i in range(len(q_est)):
    qc = Quaternion(*q_est[i, :])
    p_est_euler.append(qc.to_euler())

    # First-order approximation of RPY covariance
    J = rpy_jacobian_axis_angle(qc.to_axis_angle())
    p_cov_euler_std.append(np.sqrt(np.diagonal(J @ P_k[i, 6:, 6:] @ J.T)))

p_est_euler = np.array(p_est_euler)
p_cov_euler_std = np.array(p_cov_euler_std)

# Get uncertainty estimates from P matrix
p_cov_std = np.sqrt(np.diagonal(P_k[:, :6, :6], axis1=1, axis2=2))

titles = ['Easting', 'Northing', 'Up', 'Roll', 'Pitch', 'Yaw']
for i in range(3):
    ax[0, i].plot(range(num_gt), gt.p[:, i] - p_est[:num_gt, i])
    ax[0, i].plot(range(num_gt),  3 * p_cov_std[:num_gt, i], 'r-')
    ax[0, i].plot(range(num_gt), -3 * p_cov_std[:num_gt, i], 'r-')
    ax[0, i].set_title(titles[i], fontsize=16)
    plt.tight_layout(pad=2)
    ax[0,i].set_xlabel('No. of Measurements', fontsize=14)
ax[0,0].set_ylabel('Meters', fontsize=16)

for i in range(3):
    ax[1, i].plot(range(num_gt), \
        angle_normalize(gt.r[:, i] - p_est_euler[:num_gt, i]))
    ax[1, i].plot(range(num_gt),  3 * p_cov_euler_std[:num_gt, i], 'r--')
    ax[1, i].plot(range(num_gt), -3 * p_cov_euler_std[:num_gt, i], 'r--')
    ax[1, i].set_title(titles[i+3], fontsize=16)
    plt.tight_layout(pad=4)
    ax[1,i].set_xlabel('No. of Measurements', fontsize=14)
ax[1,0].set_ylabel('Radians', fontsize=16)

plt.show()
```
